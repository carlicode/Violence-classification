# Extracción de características en el análisis de audio

## No violence

```python
import torch
import librosa
import matplotlib.pyplot as plt

def extract_waveform(audio_file):
    audio, sr = librosa.load(audio_file)
    waveform_tensor = torch.from_numpy(audio)

    return waveform_tensor

if __name__ == "__main__":
    audio_file = '/content/drive/MyDrive/Tesis/Test/No Violence/no_violence_10.wav'

    waveform = extract_waveform(audio_file)

    fig, ax = plt.subplots()

    ax.plot(waveform.numpy())

    plt.show()
```

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

def plot_spectrogram(audio_file):
    audio, sr = librosa.load(audio_file)

    spectrogram = librosa.feature.melspectrogram(y=audio, sr=sr)

    spectrogram_db = librosa.power_to_db(spectrogram, ref=np.max)

    plt.figure(figsize=(10, 4))
    librosa.display.specshow(spectrogram_db, sr=sr, x_axis='time', y_axis='mel')
    plt.colorbar(format='%+2.0f dB')
    plt.title('Espectrograma')
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    audio_file = '/content/drive/MyDrive/Tesis/Test/No Violence/no_violence_10.wav'

    plot_spectrogram(audio_file)
```

```python
import numpy as np
import librosa
import matplotlib.pyplot as plt

def plot_spectrum(audio_file):
    audio, sr = librosa.load(audio_file)

    spectrum = np.fft.fft(audio)

    freq = np.fft.fftfreq(len(audio), d=1/sr)

    plt.figure(figsize=(10, 4))
    plt.plot(freq, np.abs(spectrum))
    plt.xlabel('Frecuencia (Hz)')
    plt.ylabel('Amplitud')
    plt.title('Espectro de Fourier')
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    audio_file = '/content/drive/MyDrive/Tesis/Test/No Violence/no_violence_10.wav'
    plot_spectrum(audio_file)
```

```python
import numpy as np
import librosa
import matplotlib.pyplot as plt

def plot_spectrum(audio_file):
    # Cargar el archivo de audio con librosa
    audio, sr = librosa.load(audio_file)

    # Calcular la transformada de Fourier
    spectrum = np.fft.fft(audio)

    # Calcular los valores de frecuencia correspondientes
    freq = np.fft.fftfreq(len(audio), d=1/sr)

    # Visualizar el espectro
    plt.figure(figsize=(10, 4))
    plt.plot(freq, np.abs(spectrum))
    plt.xlabel('Frecuencia (Hz)')
    plt.ylabel('Amplitud')
    plt.title('Espectro de Fourier')
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    # Ruta al archivo de audio
    audio_file = '/content/drive/MyDrive/Tesis/Train/Violence/violence_225.wav'

    # Graficar el espectro de Fourier
    plot_spectrum(audio_file)
```
